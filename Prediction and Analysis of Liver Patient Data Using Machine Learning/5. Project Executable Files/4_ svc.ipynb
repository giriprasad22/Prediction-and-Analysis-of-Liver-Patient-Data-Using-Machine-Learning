```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing the dataset
data = pd.read_csv('indian_liver_patient.csv')
```

```python
data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].mode()[0],inplace=True)
```

```python
x = data.drop('Dataset',axis=1)
y = data['Dataset']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x['Gender']=le.fit_transform(x['Gender'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
svm = SVC()
```

```python
svm.fit(x_train, y_train)
```

```python
y_pred_svm = svm.predict(x_test)
```

```python
accuracy_score(y_pred_svm, y_test)
```

```python
print(classification_report(y_test,y_pred_svm))
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python

parameters={
    'kernel':['rbf'],
    'C':[0.1,0.5,1.0],
    'gamma':[0.01,0.0001]
}
RCV  = RandomizedSearchCV(estimator=SVC(),param_distributions=parameters,cv=10,n_iter=4,verbose=2)
```

```python
RCV.fit(x_train,y_train)
```

```python
RCV.best_score_
```

```python
RCV.best_params_
```

```python
model2= SVC(C=0.1, gamma=0.01, kernel='rbf')

model2.fit(x_train,y_train)
```

```python
y_predict2=model2.predict(x_test)
accuracy_score(y_test,y_predict2)
```
