```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing the dataset
data = pd.read_csv('indian_liver_patient.csv')
```

```python
data.head()
```

```python
data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].mode()[0],inplace=True)
```

```python
x = data.drop('Dataset',axis=1)
y = data['Dataset']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x['Gender']=le.fit_transform(x['Gender'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.25)
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
ypred_knn = knn.predict(x_test)
```

```python
knn_acc = accuracy_score(ypred_knn, y_test)
print(knn_acc)
```

```python
print(classification_report(y_test,ypred_knn))
```

```python
confusion_matrix(y_test,ypred_knn)
```

```python
data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].median(),inplace=True)
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
x = data.drop('Dataset',axis=1)
y = data['Dataset']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm=MinMaxScaler()
```

```python
x=norm.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=28)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6, weights='uniform',
    algorithm='kd_tree',
    leaf_size=20)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
