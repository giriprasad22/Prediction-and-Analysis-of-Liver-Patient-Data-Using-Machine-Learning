```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing the dataset
data = pd.read_csv('indian_liver_patient.csv')
```

```python
data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].mode()[0],inplace=True)
```

```python
x = data.drop('Dataset',axis=1)
y = data['Dataset']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x['Gender']=le.fit_transform(x['Gender'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train, y_train)
```

```python
ypred_rfc = rfc.predict(x_test)
```

```python
rfc_acc = accuracy_score(ypred_rfc, y_test)
print(rfc_acc)
```

```python
print(classification_report(y_test,ypred_rfc))
```

```python
confusion_matrix(y_test,ypred_rfc)
```
